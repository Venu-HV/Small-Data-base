```python
from numpy import average
import pandas as pd
import matplotlib.pyplot as plt
subjects=[
    "Kannada",
    "English",
    "Hindi",
    "Mathematics",
    "science",
    "social science"
]
marks=[102,88,79,74,87,100]
df=pd.DataFrame({"subjects":subjects,"marks":marks})
print("10th Marks Analysis\n")
print(df)
Total=df["marks"].sum()
average=df["marks"].mean()
print("\nTotal Marks:",Total)
print("Average Marks:",round(average,2))

top=df.loc[df["marks"].idxmax()]
print("Highest Scoring Subject:",top["subjects"])

lowest=df.loc[df["marks"].idxmin()]
print("Lowest Scoring Subject:",lowest["subjects"])

plt.pie(
    df["marks"],
    labels=df["subjects"],
    autopct="%1.1f%%"
)
plt.title("10th Marks Analysis")
plt.show()
```
